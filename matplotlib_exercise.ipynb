1. Use matplotlib to plot the following equation:
    
        y = x^2 - x + 2

        * You'll need to write the code that generates the x and y points.
        * Add an anotation for the point 0, 0, the origin.

```python
#Generate x and y points
x = range(-100, 100)
y = [(num**2 - num + 2) for num in x]
```

```python
# Plot x and y with plot()
plt.plot(x,y)

# Add title
plt.title('$y = x^2 - x +2$')

# Add x and y labels
plt.xlabel('$x$')
plt.ylabel('$y$')

# Annotate (0,0) with coordinate
plt.annotate('Origin', xy = (0,0), xytext = (0, 3000), arrowprops ={})

# Set y limit
plt.ylim(0,10500)

plt.show()
```

2. Create and label 4 separate charts for the following equations (choose a range for x that makes sense):
    
        y = √(x)
        y = x^3
        y = 2^x
        y = 1/(x+1)

* You can use functions from the math module to help implement some of the equations above.

```python
# Generate x1 and y1 values

x1 = range(0,100)
y1 = [math.sqrt(n) for n in x1]

# plot the data
plt.plot(x1,y1)

# Customize
plt.title('$ y = \sqrt{x}$')
plt.xlabel('$x$')
plt.ylabel('$\sqrt{x}$')
plt.xlim(0,100)
plt.ylim(0, 10)

plt.grid(True, ls = '--');

plt.show()
```

```python
# Generate x2 and y2 values
x2 = range(0,20)
y2 = [n**3 for n in x2]

#plot data
plt.plot(x2,y2)

# Customize
plt.title('$ y = x^3$')
plt.xlabel('$x$')
plt.ylabel('$x^3$')
plt.xticks(range(0,20,5))
plt.xlim(0,20)
plt.ylim(0)
plt.grid(True, ls = '--');
```

```python
# Generate x3 and y3 values
x3 = [x * 0.1 for x in range(0,150)]
y3 = [2 ** n for n in x3]

# plot data
plt.plot(x3,y3)

# Customize
plt.title('$ y = 2^x$')
plt.xlabel('$x$')
plt.ylabel('$2^x$')
plt.grid(True, ls = '--')

# Add ticks as well
plt.xticks(range(0, 16));

```

```python
# Note 0.1x is used for a smoother curve
# x4 and y4 
x4 = [x * 0.1 for x in range(0,150)]
y4 = [1/(n+1) for n in x4]


# Plot the data
plt.plot(x4,y4)


# Customize
# Add title
plt.title('$ y = \dfrac{1}{x+1}$')

# Add x and y labels
plt.xlabel('$x$')
plt.ylabel('$\dfrac{1}{x+1}$')


plt.grid(True, ls = '--')

plt.show()
```

3. Combine the four figures from exercises 2 into one large figure using subplots. 

```python

# Set figure size for 4 subplots
plt.figure(figsize = (10,6))

# Subplot 1
plt.subplot(2,2,1)

plt.plot(x1,y1)

plt.title('$ y = \sqrt{x}$')
plt.xlabel('$x$')
plt.ylabel('$\sqrt{x}$')
plt.xlim(0,100)
plt.ylim(0, 10)
plt.grid(True, ls = '--')

#subplot 2
plt.subplot(222)

plt.plot(x2,y2)

plt.title('$ y = x^3$')
plt.xlabel('$x$')
plt.ylabel('$x^3$')
plt.xticks(range(0,20,5))
plt.xlim(0,20)
plt.ylim(0)
plt.grid(True, ls = '--')


#subplot 3
plt.subplot(223)

plt.plot(x3,y3)
plt.grid(True, ls = '--')
plt.xlabel('$x$')
plt.ylabel('$2^x$')
plt.title('$y = 2^x$')

#subplot 4
plt.subplot(224)

plt.plot(x4,y4)
plt.title(r'$ y = \frac{1}{x+1}$')
plt.xlabel('$x$')
plt.ylabel(r'$\frac{1}{x+1}$')
plt.grid(True, ls = '--')
plt.xlim(0,15)
plt.ylim(0)


plt.tight_layout()

```

4. Combine the four figures from exercises 2 into one plot. Include a legend to distinguish each equation. 

```python
# Set figure size
plt.figure(figsize = (8,6))

# Data is the same as in exercise 2

# Plot the data
# Include a color and label
# Add transparency

plt.plot(x1, y1, c='#30a2da', label='$\sqrt{x}$', alpha=1)
plt.plot(x2, y2, c='orange', label='$x^{3}$', alpha=1)
plt.plot(x3, y3, c= (1,0,1), label='$2^{x}$', alpha=1)
plt.plot(x4, y4, c= '#444444', label='$1/(x+1)$', alpha=1)

# Adjust xlim and ylim
plt.ylim(0, 10)
plt.xlim(0, 10)

plt.ylabel('$y$')
plt.xlabel('$x$')
plt.title('Combined Charts')
plt.legend()

plt.show()

```
