````python
import matplotlib.pyplot as plt
import math as math
import numpy as np
import pandas as pd
````

<b>Example 1:

````python
x= (range(1,20))
# Curves for the graph
y1 = [1 for n in x]
y2 = [math.log(n) for n in x]
y3 = x
y4= [(n * math.log(n)) for n in x]
y5= [n**2 for n in x]
y6= [2**n for n in x]
y7= [math.factorial(n) for n in x]
y8= [n**n for n in x]

plt.figure(figsize = (18,8))

plt.plot(x,y1, color = 'magenta', label= 'y1= $O(1)$', alpha = 1)
plt.plot(x,y2, color = 'maroon', label= 'y2 = $O(log n)$', alpha = 1)
plt.plot(x,y3, color = 'red', label= 'y3= $O(n)$', alpha = 1)
plt.plot(x,y4, color= 'purple', label= 'y4 = $O(n log n)$', alpha = 1)
plt.plot(x,y5, color= 'blue', label= 'y5 = $O(n^2)$', alpha = 1)
plt.plot(x,y6, color= 'green', label= 'y6 = $O(2^n)$', alpha = 1)
plt.plot(x,y7, color= 'cyan', label= 'y7 = $O(n!)$', alpha = 1)
plt.plot(x,y8, color= 'brown', label= 'y8 = $O(n^n)$', alpha = 1)

# Title for the chart 
plt.title('Big O Notation', fontsize = 20)
# Label for x axis
plt.xlabel('Elements')
# Label for y axis
plt.ylabel('Operations')

plt.annotate('LaTex', xy= (1.5,10))

#Legend for the chart
plt.legend(loc = 'upper right', fontsize = 14)
plt.xlim(1, 11)
plt.ylim(0, 50)

plt.show()
````
