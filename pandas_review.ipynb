Columns

```python
mpg.head()
```

```python
mpg = mpg.rename(columns = {'displ': 'displacement',
                            'cyl': 'cylinders',
                            'drv': 'drive'})

mpg.head()
```

```python
mpg = mpg.drop(columns = ['displacement', 'cylinders'])

mpg.head()
```

```python
mpg = mpg.assign(displacement = mpg.hwy - mpg.cty)

mpg.head()
```

Plotting

```python
plt.plot(mpg.cty)
```

```python
plt.scatter(mpg.cty, mpg.hwy)
plt.xlabel('City MPG')
plt.ylabel('Highway MPG')
```

```python
plt.hist(mpg.hwy)
```

```python
grpby = mpg.groupby('drive').hwy.mean()
grpby
```

```python
plt.bar(grpby.index, grpby)
```
